```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figure=(15,10))
plt.hist(data_new['Fat'], bins=10,edgecolor='red')
plt.title('Fat Range')
plt.xlabel('Fat')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
plt.figure(figure=(10,3))
sns.histplot(data_new['Fat'], bins=5 ,kde=True)
plt.title('Fat Range')
plt.xlabel('Fat')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
bmi_plot=data_new.groupby('Date')['BMI'].mean()
plt.figure(figsize=(20,6))
bmi_plot.plot()
plt.title('BMI-Plot')
plt.xlabel('Date')
plt.ylabel('BMI')
plt.show()
```

```python
cor=data_new['Fat'].corr(data_new['BMI'])
plt.figure(figsize=(10,6))
plt.scatter(data_new['Fat'],data_new['BMI'])
plt.title('Relation between BMI and Fat')
plt.xlabel('Fat')
plt.ylabel('BMI')
plt.show()
```

```python
healthy_BMI= data_new[data_new['BMI']<25]
unhealthy_BMI = data_new[data_new['BMI']>=25]
plt.figure(figsize=(10,6))
sns.histplot(healthy_BMI['Fat'],bins=20,kde=True, alpha=0.5, label='Healthy BMI')
sns.histplot(unhealthy_BMI['Fat'],bins=20, kde=True, alpha=0.5,label='Unhealthy BMI')
plt.title('Fat percent of healthy and unhealthy BMI people')
plt.xlabel('BMI')
plt.ylabel('frequency')
plt.legend()
plt.grid()
plt.show()
```

```python
labels=['HealthyBMI',' UnhealthyBMI']
data_new['BMI'] = pd.to_numeric(data_new['BMI'], errors='coerce')
healthy_BMI= data_new[data_new['BMI']<25].shape[0]
unhealthy_BMI = data_new[data_new['BMI']>=25].shape[0]
attributes=[healthy_BMI,unhealthy_BMI]
plt.figure(figsize=(8,6))
plt.pie(attributes,labels=labels)
plt.title('BMI Distribution')
plt.legend()
plt.show()
```
